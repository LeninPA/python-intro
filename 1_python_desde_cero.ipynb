Las listas guardan elementos manteniendo un orden entre ellos. Se pueden almacenar elementos de cualquier tipo. Cada elemento tiene un "índice" que indica la posición del elemento dentro de la lista, empezando desde 0.

Para crear una lista basta con poner cosas entre [] (corchetes) y separarlas con , (comas).

```python
print("Mi lista:")

lst = ["Sonic",1,"Tails","Knuckles"] #Lista con tres strings y un número
print(lst)

#La función len me da la longitud de la lista
print("Longitud de la lista:")
n = len(lst)
print(n)

#Para encontrar en que posición se encuentra un elemento 
print("Indice de Knuckes:")
print(lst.index("Knuckles"))


# Acceder a un elemento (indexación)
print ("Tercer elemento:")
print (lst [#Completa])

print ("Último elemento:")
print (lst [#Completa])
```

```python
#Añadir elementos
lst1 = ["Sonic", 1, "Tails", "Knuckles"]
lst1.append(3) #Se añade al final de la lista
print(lst1)

#Remover n elemento
lst1.remove ("Tails") #Remueve por valor, pero solo la primera recurrencia
print(lst1)

#Remover por posición o índice
lst1.pop(2)
print(lst1)

#Reemplazar elementos ya existentes
lst1[1] = 10
print(lst1)


#Ver si un elemento esta en una lista
print("Knuckles" in lst)

#Verifica si "Sega" esta o no en la lista
```

También Python tiene ciertas funciones que otros lenguajes no:

```python
#Ordenar una lista de forma ascendente
lst2 = ["Ramos", "Ross", "Romita", "Campbell", "Crain", "Quesada"]
lst2.sort()
print(lst2)

#Ordenar una lista de forma descendente
lst2.reverse()
print(lst2)
```

La forma en que "recorreremos" (visitamos valor por valor) una colección es la siguiente:

```python
for elem in lst2:
    print(elem) #Por cada iteración del lazo, elem va a tomar como valor cada uno de los elementos de la lista
```

Ahora, como concepto adicional:

```python
for i in range (len(lst2)):
    print(i) #Recorrer por indice
    
for i in range (len(lst2)):
    print(lst2[i]) #Recorrer por índice y luego acceder al valor
    
for elem in lst2:
    print(elem) #Recorrer por valores (forma más simple)
```

## Conjuntos

Los conjutnos también son una colección pero se definen entre {} (llaves) estos no mantienen un orden y por lo tanto no son estructuras indexables, no almacenan elementos repetidos. Podemos aplicar sobre ellos las operaciones matemáticas clásicas de los conjuntos.

```python
#Definimos un conjunto
cjnt = {1, 2, 4, 6, 7}

#De lista a conjunto
lst4 = (14, 3, 8, 7, 9 , 11)
cjnt1 = set (lst4)
print(cjnt1)

#Añadimos elementos
cjnt.add(19)
print(cjnt) #Notamos que no guardan orden

#Añade 22 como elemento


cjnt.add(6)
print(cjnt) #No guarda repetidos

#Remover elemento
cjnt.remove(19)
print(cjnt)

#Recorrer conjuntos
for elemen in cjnt:
    print(elemen)
    
#Existencia de un elemento
print( 2 in cjnt)

#Verifica que 8 este en el conjunto y que lo muestre
  

#Unión
print(cjnt.union(cjnt1))

#Intersección
print(cjnt.intersection(cjnt1))
```

## Diccionarios

Colecciones donde cada elemento que se almacena es un par. Su sintáxis es la siguiente:


```python
{"edad" : 20,
"nombre" : Linda} #clave : valor
```

Los diccionarios tampoco mantienen orden entonces tampoco hay índices, el acceso a los valores (indexación) se realiza a través de las claves.

```python
dicc =  {
    "nombre": "Peter",
    "edad": 25,
    "compañia" : "Parker Industries"
}
print(dicc["edad"]) #Aqui obtenemos el valor "indexado" con su clave
```

```python
# Añadir elementos
dicc["apellido"] = "Parker"
print(dicc)

#Añade el elemento "alias" al diccionario y asignale "Spiderman" como valor


#Editar un valor reemplazando
dicc["edad"] = 31
print(dicc)

#Editar un valor con operaciones
dicc["edad"] -= 6
print(dicc)

#Eliminar elementos
del dicc["compañia"]
print(dicc)

#Elimina el elemento "alias" para proteger la identidad del héroe


#Recorrer un diccionario
#el lazo for tendra una sintaxis diferente
for h,k in (dicc.items()): #Aqui lo trasformo en una lista 
    print(h,k)

```

Nota: Puedo crear también estructuras anidadas. Como ejemplo, en un diccionario como valor puedo tener un arreglo on incluso otro diccionario.

Arreglos que nos seran muy utiles son las matrices y vectores. Un vector es simplemente un arreglo unidimensional, pudiendo ser representado como una lista (sin usar numpy).

```python
# Vector fila
## COMPLETA CON 4 ELEMENTOS
v = []

print(v)

# Acceso a un elemento
print(v[])   # completa para el 1er elemetno 
print(v[])   # completa para el 3er elemento
```

Mientras que una matriz es un arreglo bidimensional 2D y se puede representar como una lista que contiene otras listas (cada sublista es una fila):

```python
# Matriz 2x3
M = [
    [1, 2, 3],
    [4, 5, 6]
]

# Acceso a un elemento: fila 1, columna 2
print(M[0][1])  # 2
```

Algunas operaciones que podremos realizar con estos arreglos son:

```python
# Suma de vectores
u = [1, 1, 1]
v = [1, 2, 3]

suma = [u[i] + v[i] for i in range(len(u))]
print(suma)   # [2, 3, 4]

# Producto escalar (dot product)
dot = sum(u[i] * v[i] for i in range(len(u)))
print(dot)    # 6

# Producto matriz-vector
M = [
    [1, 2, 3],
    [4, 5, 6]
]
v = [1, 2, 3]

resultado = []
for fila in M:
    suma = 0
    for j in range(len(v)):
        suma += fila[j] * v[j]
    resultado.append(suma)

print(resultado)  # [14, 32]

# Transpuesta
M = [
    [1, 2, 3],
    [4, 5, 6]
]

# Transpuesta
T = [[M[j][i] for j in range(len(M))] for i in range(len(M[0]))]

print(T)
# Resultado: [[1, 4], [2, 5], [3, 6]]
```

## Errores (Sencillos)

A veces podemos "blindar" el programa en caso que halla errores en pedazos con las instrucciones try y except para que el programa no se caiga, si no que continue con el flujo. Ejemplo

```python
#Programa que va a brincar error ¿Por que?
cdn =  "Oh no otra cadena"
x = int(cdn)
print("meh")
```

```python
cdn =  "Oh no otra cadena"
try: 
    x = int(cdn)
except: 
    print ("Tenemos un error")
    
print("meh")
```

```python
#Para obtener el error
cdn =  "Oh no otra cadena"
try: 
    x = int(cdn)
except Exception as err: 
    print ("Hay un error", err)
    
print("meh")
```
