### Otras funciones


#### Algebra Lineal

```python

from numpy import linalg as LA

    #Eigenvalores y eigenvectores generales
w, v = LA.eig(np.diag((1, 2, 3)))
w; v
    #Valores propios de una matriz no simétrica
D = np.diag((-1,1))
LA.eigvals(D)

    #Matriz o vector norma
a = np.arange(9) - 4
LA.norm(a)

    #Determinante de una matriz
a = np.array([[1, 2], [3, 4]])
np.linalg.det(a)
    #Determinante de un conjunto de matrices
a = np.array([ [[1, 2], [3, 4]], [[1, 2], [2, 1]], [[1, 3], [3, 1]] ])
a.shape
np.linalg.det(a)
```

```python
#Inverso multiplicativo de una matriz

from numpy.linalg import inv
a = np.array([[1., 2.], [3., 4.]])
inv(np.matrix(a))

    #De varias matrices
c = np.array([[[1., 2.], [3., 4.]], [[1, 3], [3, 5]]])
inv(c)
```

```python
#Resolver una ecuación de matriz lineal o un sistema de ecuaciones escalares lineales
a = np.array([[1, 2], [3, 5]])
b = np.array([1, 2])
x = np.linalg.solve(a, b)
x
```

```python
#Checar si la solución es correcta
np.allclose(np.dot(a, x), b)
```

```python
#Solución de mínimos cuadrados a una ecuación de matriz lineal
    #Sea y = mx + c, y :
x = np.array([0, 1, 2, 3])
y = np.array([-1, 0.2, 0.9, 2.1])
    #Podemos reescribir la ecuación lineal como y = Ap, donde A = [[x 1]] y p = [[m], [c]]
A = np.vstack([x, np.ones(len(x))]).T
A
```

```python
    #Ahora usa lstsq para resolver p:
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
m,c
```

```python
#Aqui ajuste una línea, y = mx + c, a través de la muestra de puntos de datos ruidosos:
import matplotlib.pyplot as plt
_ = plt.plot(x, y, 'o', label='Original data', markersize=10)
_ = plt.plot(x, m*x + c, 'r', label='Fitted line')
_ = plt.legend()
plt.show()
```

#### Subsetting, Slicing e Indexación

```python
#Subsetting

a[2] #Selecciona el elemento con el indice 2

a[0] # #Selecciona el elemento con el indice 0

#Seleccione el elemento en la fila 1, columna 2 (equivalente a b [1] [2])
b = np.array([[ 1.5, 4. , 9. ],
             [ 4. , 10. , 18. ]])
b[1,1] 
```

```python
#Slicing

b[0:2] #Seleccionar elementos en el índice 0 y 1

b[0:2,1]#Seleccionar elementos en las filas 0 y 1 en la columna 1

b[:1] #Seleccionar todos los elementos de la fila 0 (equivalente a b [0: 1,:])

b[1,...] #Equivalente a [1,:,:]

b[ : :-1]#Array en reversa

b[[1, 0, 1, 0],[0, 1, 2, 0]]#Selecciona elementos (1,0),(0,1),(1,2) y (0,0)

b[[1, 0, 1, 0]][:,[0,1,2,0]]#Selecciona un subconjunto de filas y columnas de la matriz.
```

```python
# Boolen Slicing

b[b<2] #Selecciona elementos menores a 2
```
